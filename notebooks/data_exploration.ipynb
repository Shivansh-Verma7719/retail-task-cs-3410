```python
def one_hot_encode(X):
    print(f"\nStarting one-hot encoding...")
    
    # Identify categorical columns (object or boolean type)
    categorical_cols = X.select_dtypes(include=['object', 'boolean']).columns.tolist()
    
    print(f"Found {len(categorical_cols)} categorical columns: {categorical_cols}")
    
    if not categorical_cols:
        print("No categorical columns found, returning original dataframe")
        return X
    
    X_encoded = X.copy()
    
    # Set a threshold for cardinality
    max_unique_threshold = 10
    
    encodable_cols = [col for col in categorical_cols if X_encoded[col].nunique() <= max_unique_threshold]
    high_cardinality_cols = [col for col in categorical_cols if col not in encodable_cols]
    
    print(f"Columns suitable for one-hot encoding ({len(encodable_cols)}): {encodable_cols}")
    print(f"High cardinality columns to drop ({len(high_cardinality_cols)}): {high_cardinality_cols}")
    
    # Drop high cardinality columns
    if high_cardinality_cols:
        X_encoded = X_encoded.drop(columns=high_cardinality_cols)
        print(f"Dropped columns: {high_cardinality_cols}")
        
    # Apply one-hot encoding using pandas' get_dummies
    if encodable_cols:
        # get_dummies can process all columns at once, which is more efficient
        X_encoded = pd.get_dummies(X_encoded, columns=encodable_cols, drop_first=True)
    
    print(f"\nOne-hot encoding completed!")
    print(f"Original shape: {X.shape}, Encoded shape: {X_encoded.shape}")
    
    return X_encoded
```

```python
X = df.drop('avg_purchase_value', axis=1)
y = df['avg_purchase_value']

# Apply the one-hot encoding
X_processed = one_hot_encode(X)

print("\n--- Processed Features (X) after One-Hot Encoding ---")
print(X_processed.head())
```

```python
final_df = pd.concat([X_processed, y], axis=1)

correlation_matrix = final_df.corr()

# Visualize the correlation matrix using a heatmap
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix, 
    annot=False,          # Show the correlation values on the heatmap
    cmap='coolwarm',     # Use a color map that's good for correlations (positive/negative)
    fmt=".2f",           # Format the numbers to two decimal places
    linewidths=.5
)
plt.title('Correlation Matrix of Features and Target', fontsize=16)
plt.show()
```
